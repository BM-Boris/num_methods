Метод Зейделя


```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
from mpl_toolkits.mplot3d import Axes3D
```

```python
e=0.001
n=11
z=np.zeros((n,n))
h=0.1
hh=0.1
u=4
x=[i*h for i in range(n)]
y=[i*hh for i in range(n)]
xgrid, ygrid=np.meshgrid(x, y)
xx=np.arange(0,1.1,0.1)
yy=np.arange(0,1.1,0.1)
xx,yy=np.meshgrid(xx,yy)

for i in range(n):
    z[i][0]=y[i]**2
    z[i][n-1]=1+x[i]**2
    z[0][i]=y[i]**2
    z[n-1][i]=1+x[i]**2
  
def zei(z):
    for i in range(1,n-1):
        for j in range(1,n-1):
            z[i][j]=(hh**2*z[i+1][j]+hh**2*z[i-1][j]+h**2*z[i][j+1]+h**2*z[i][j-1]-u*h**2*hh**2)/(2*(h**2+hh**2))
    return z

while np.linalg.norm(z-(xx**2+yy**2))>e: 
    z=zei(z)
    
fig=pylab.figure()
axes=Axes3D(fig)
axes.plot_surface(xgrid,ygrid,z)
plt.title("Seidel method")
fig=plt.figure()
axes=Axes3D(fig)
plt.title("Exact solution")
axes.plot_surface(xx,yy,(xx**2+yy**2))
pylab.show()
```
