```python
import matplotlib.pyplot as plt
for a in range(0,10):
  fi = np.linspace(-2*np.pi,2*np.pi,100)
  r = a + np.cos(fi)
  x = r * np.cos(fi)
  y = r * np.sin(fi)
  plt.plot(x,y)
```
